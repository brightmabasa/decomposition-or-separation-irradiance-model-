```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("DIF_RMSE_ALL.csv")

# Extract and sort RMSE values
model_rmse = df.iloc[0]
sorted_rmse = model_rmse.sort_values()

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(sorted_rmse.index, sorted_rmse.values, color='grey', width=0.5)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.ylabel('RMSE (W/m²)', fontsize=14)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the figure
plt.savefig("DIF_RMSE_ALLSKY.png", dpi=300)

# Display the plot
plt.show()

```
